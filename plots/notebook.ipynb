```python
import numpy as np 
import matplotlib.pyplot as plt 
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
# Arquivos de dados
base_filename = "datafiles/geratriz_"
ext_file = ".txt"
geratrizes_files = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M"]
files = [base_filename + f + ext_file for f in geratrizes_files]
```

```python
# Ler os dados
data = {}
for i in range(len(files)):
    data[geratrizes_files[i]] = np.loadtxt(files[i], unpack=False,)
vec_data = list(data.values())
```

```python
# Plots para A, B, C
fig, ax = plt.subplots()
plots_tri = ("A", "L", "M")
for index in plots_tri:
    ax.plot(data[index][:,0], data[index][:,1], '-o', label=index)

ax.legend()
ax.grid()
ax.set_xlim(-.01,0.115)
ax.set_ylim(15,160)
ax.set_xlabel("Posição")
ax.set_ylabel("T [ºC]")
plot_name = plots_tri[0] + plots_tri[1] + plots_tri[2] + ".svg"
plt.savefig(plot_name)
```

```python
# Diagrama de barras para a taxa
fig, ax = plt.subplots()
q_dot = np.array([data[index][:,3][0] for index in geratrizes_files])
q_x = np.array([i for i in range(1, len(geratrizes_files)+1)])
ax.bar(q_x, q_dot, tick_label=geratrizes_files)
ax.set_xlabel("Geratriz")
ax.set_ylabel(r'$\dot{q}$ [W]')
plt.savefig("taxas.svg")

```

```python
# Diagrama de barras para o volume
fig, ax = plt.subplots()
q_dot = np.array([data[index][:,4][0]*1e6 for index in geratrizes_files])
q_x = np.array([i for i in range(1, len(geratrizes_files)+1)])
ax.bar(q_x, q_dot, tick_label=geratrizes_files)
ax.set_xlabel("Geratriz")
ax.set_ylabel(u'Volume [$cm^3$]')
plt.savefig("volumes.svg")
```
