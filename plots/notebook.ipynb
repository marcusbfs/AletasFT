```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Arquivos de dados
base_filename = "datafiles/geratriz_"
ext_file = ".txt"
geratrizes_files = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M"]
files = [base_filename + f + ext_file for f in geratrizes_files]
```

```python
# Ler os dados
data = {}
for i in range(len(files)):
    data[geratrizes_files[i]] = np.loadtxt(files[i], unpack=False, usecols=(0,1))
vec_data = list(data.values())
```

```python
# Plots para A, B, C
fig, ax = plt.subplots()
for index in ("A", "B","C"):
    ax.scatter(data[index][:,0], data[index][:,1], label=index)

ax.legend()
ax.grid()
ax.set_xlim(-.01,0.115)
ax.set_ylim(15,160)
```
